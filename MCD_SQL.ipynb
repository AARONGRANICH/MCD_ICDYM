Iniciamos Importando las Librerias necesarias

```python
import pymysql
import matplotlib.pyplot as plt
import pandas as pd
```

Generamos la coneccion con la libreria de pymysql 

```python
connection = pymysql.connect(host='localhost',
                                 user='root',
                                 password='YourPassword',                             
                                 db='NFL_MCD')
```

Vaoms a revisar el DF por medio de un SQL Query

```python
dfr=pd.read_sql_query("SELECT * FROM NFL_MCD.baseballdata", connection, index_col='MyUnknownColumn')
dfr.head(10)
```

Generamos un nuevo DF

```python
Wins= dfr.groupby("Tm").sum()["W"].reset_index()
Wins
```
